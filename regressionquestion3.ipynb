```python
import numpy as np 
import matplotlib.pyplot as plt
import numpy.random as rnd
import pandas as pd
from pandas import DataFrame
import random as random 
import sklearn as sk
from sklearn.linear_model import LinearRegression

#load all the data
data = pd.read_csv("autoData.csv", header=None)
cylinders = data.iloc[:,2].copy()
displacement = data.iloc[:,3].copy()
horsepower = data.iloc[:,4].copy()
acceleration = data.iloc[:,5].copy()
weight = data.iloc[:,6].copy()
mpg= data.iloc[:,7].copy()

#for part a: cylinders vs mpg
#reshape cylinders to a 2D array
cylinder_array = np.array(cylinders.values.tolist())
cylinder_reshape = cylinder_array.reshape((-1,1))

#use the regression model to find the line of best fit 
reg = LinearRegression().fit(cylinder_reshape,mpg)
r_sq = reg.score(cylinder_reshape, mpg)
#plot the points 
plt.plot(cylinders, reg.coef_*cylinders + reg.intercept_,color='red')
plt.scatter(cylinders,mpg)
plt.title('Linear Regression of Cylinders vs MPG')
plt.xlabel('Cylinders')
plt.ylabel('mpg')
plt.show()

print('slope:',reg.coef_)
print('intercept:',reg.intercept_)
print('coefficient of determination:',r_sq)

#for part b: displacement vs mpg 
displacement_array=np.array(displacement.values.tolist())
displacement_reshape = displacement_array.reshape((-1,1))

reg_1 = LinearRegression().fit(displacement_reshape,mpg)
r_sq1 = reg_1.score(displacement_reshape, mpg)

plt.plot(displacement,reg_1.coef_*displacement + reg_1.intercept_,color='red')
plt.scatter(displacement,mpg)
plt.title('Linear Regression of Displacement vs MPG')
plt.xlabel('Displacement')
plt.ylabel('mpg')
plt.show()

print('slope:',reg_1.coef_)
print('intercept:',reg_1.intercept_)
print('coefficient of determination:',r_sq1)

#for part c: horsepower vs mpg 
horsepower_array=np.array(horsepower.values.tolist())
horsepower_reshape =horsepower_array.reshape((-1,1))

reg_2 = LinearRegression().fit(horsepower_reshape,mpg)
r_sq2 = reg_2.score(horsepower_reshape, mpg)

plt.plot(horsepower,reg_2.coef_*horsepower + reg_2.intercept_,color='red')
plt.scatter(horsepower,mpg)
plt.title('Linear Regression of Horsepower vs MPG')
plt.xlabel('Horsepower')
plt.ylabel('mpg')
plt.show()

print('slope:',reg_2.coef_)
print('intercept:',reg_2.intercept_)
print('coefficient of determination:',r_sq2)

#for part d: weight vs mpg 
weight_array=np.array(weight.values.tolist())
weight_reshape =weight_array.reshape((-1,1))

reg_3 = LinearRegression().fit(weight_reshape,mpg)
r_sq3 = reg_3.score(weight_reshape, mpg)

plt.plot(weight,reg_3.coef_*weight + reg_3.intercept_,color='red')
plt.scatter(weight,mpg)
plt.title('Linear Regression of Weight vs MPG')
plt.xlabel('Weight')
plt.ylabel('mpg')
plt.show()

print('slope:',reg_3.coef_)
print('intercept:',reg_3.intercept_)
print('coefficient of determination:',r_sq3)

#for part e: acceleration vs mpg 
acceleration_array=np.array(acceleration.values.tolist())
acceleration_reshape =acceleration_array.reshape((-1,1))

reg_4 = LinearRegression().fit(acceleration_reshape,mpg)
r_sq4 = reg_4.score(acceleration_reshape, mpg)

plt.plot(acceleration,reg_4.coef_*acceleration + reg_4.intercept_,color='red')
plt.scatter(acceleration,mpg)
plt.title('Linear Regression of Acceleration vs MPG')
plt.xlabel('Acceleration')
plt.ylabel('mpg')
plt.show()

print('slope:',reg_4.coef_)
print('intercept:',reg_4.intercept_)
print('coefficient of determination:',r_sq4)
```
